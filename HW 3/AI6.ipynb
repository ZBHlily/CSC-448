<b> Question:   
    - read the "https://raw.githubusercontent.com/data-8/textbook/main/assets/data/san_francisco_2015.csv" data   
    - get the info and describe of the data frame    
    - print out the size of the data frame     
    - print out the columns of the dataframe    
    - sort the dataframe based on the 'Salaries'   
    - print out the 40 percentage value of "Salaries"   
    - resample 1000 from the 'Total Compensation'> 9000 and get the median of the resample  data   
    - repeate the last step 10 times and get the average median of these resamples  
    - ensclupe the bootstrap steps into a function

```python
import pandas as pd
import numpy as np

# Function to encapsulate the bootstrap steps
def bootstrap(df, num_iterations):
    median_values = []
    for _ in range(num_iterations):
        resample = df[df['Total Compensation'] > 9000].sample(n=1000, replace=True)
        median_values.append(resample['Total Compensation'].median())
    return np.mean(median_values)

# Read the data
url = "https://raw.githubusercontent.com/data-8/textbook/main/assets/data/san_francisco_2015.csv"
df = pd.read_csv(url)

# Get info and description of the dataframe
print("Info of the DataFrame:")
print(df.info())
print("\nDescription of the DataFrame:")
print(df.describe())

# Print the size of the DataFrame
print(f"Size of the DataFrame: {df.shape}")

# Print the columns of the DataFrame
print(f"Columns of the DataFrame: {df.columns}")

# Sort the DataFrame based on 'Salaries'
df_sorted = df.sort_values('Salaries')

# Print out the 40th percentile value of "Salaries"
percentile_40 = np.percentile(df['Salaries'], 40)
print(f"40th percentile value of 'Salaries': {percentile_40}")

# Bootstrap 10 times and get the average median of these resamples
average_median = bootstrap(df, 10)
print(f"Average median of resamples: {average_median}")
```

```python
class ApplePieRecipe:
    def __init__(self, ingredient1, ingredient2):
        self.ingredient1 = ingredient1
        self.ingredient2 = ingredient2

    def method1(self, parameter1):
        return 1

# Creating an instance of the ApplePieRecipe class
applepie1 = ApplePieRecipe("sugar", "water")
ret1 = applepie1.method1("dsf")

# Creating another instance of the ApplePieRecipe class
applepie2 = ApplePieRecipe("honey", "milk")

# Printing the attributes of the instances
print(f"Ingredients of applepie1: {applepie1.ingredient1}, {applepie1.ingredient2}")
print(f"Ingredients of applepie2: {applepie2.ingredient1}, {applepie2.ingredient2}")
```
